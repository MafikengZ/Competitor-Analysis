#### 2.4. Scrolling to the bottom of our competitors profile page

```python
SCROLL_PAUSE_TIME = 5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

## 3. Extracting data from my competitors profile

#### 3.1 Declaring my variables for data storage

```python
post_date = []
post_caption = []
post_types = []
post_reaction_count = []
post_comments_count = []
post_share_count = []
```

#### 3.2 Extracting the date of the posts

```python
date_elements = '//div/a//div//span[3]/span/span[1]'
dates = driver.find_elements_by_xpath(date_elements)
for date in dates:
    post_date.append(date.text)

print(len(post_date))
```

#### 3.2.1 printing our post date data

```python
print(post_date)
```

#### 3.3 Extracting the captions of the post

```python
caption_element = '//div//div[4]/div/div/span/span/span'
captions = driver.find_elements_by_xpath(caption_element)

for caption in captions:
    post_caption.append(caption.text)

# printing the length of our captions data
print(len(post_caption))
```

##### 3.3.1 Printing our captions data

```python
print(post_caption)
```

#### 3.4 Extracting and categorizing the type of post

```python
image = '//div//div[5]/div/div/button/div/div/img'
article = 'article/div/div/a'
external_video_youtube = '//div//div[5]/div/div/div/a'
internal_video_linkedin = '//div/div/div/div/div/video'
pdf = '//div/div/iframe'


# imgs = '//div/div/div/button/div/div/img'

post_type_a = driver.find_elements_by_xpath(image)
post_type_b = driver.find_elements_by_xpath(article)
post_type_c = driver.find_elements_by_xpath(external_video_youtube)
post_type_d = driver.find_elements_by_xpath(internal_video_linkedin)
post_type_e = driver.find_elements_by_xpath(pdf)


for post_a in post_type_a:
    post_types.append("image")
for post_b in post_type_b:
    post_types.append("article")
for post_c in post_type_c:
    post_types.append("youtube video")
for post_d in post_type_d:
    post_types.append("linked video")
for post_e in post_type_e:
    post_types.append("a pdf file")
post_f = ""
if post_f != post_a :
    post_types.append("Not a formal post")
print(post_types)
```

#### 3.5 Extracting reactions counts

```python
reactions_element = '//ul/li[1]/button/span/span[1]'
reactions = driver.find_elements_by_xpath(reactions_element)
for reaction in reactions:
    post_reaction_count.append(reaction.text)

print(post_reaction_count)
```

#### 3.6 Extracting comments counts

```python
comments_element = '//div/div/div/div/div[6]//li[2]/button/span'
comments = driver.find_elements_by_xpath(comments_element)
for comment in comments:
    post_comments_count.append(comment.text)
    
print(post_comments_count)
```

#### 3.7 Extracting post share count

```python
post_share_element = '//div/ul//li[3]/div/span'
shares = driver.find_elements_by_xpath(post_share_element)
for share in shares:
    post_share_count.append(share.text)

print(post_share_count)
```

## 4. Creating a Dictionary for the data
Creating a dictionary for our data by defining it in a Key and value pair

#### 4.1 Data Dictionary declaration

```python
# creating a dictionary for our data
data = {
    "date_of_post": post_date,
    "post_captions": post_caption,
    "types_of_post": post_types,
    "reactions": post_reaction_count,
    "comments": post_comments_count,
    "shared_count": post_share_count
}

print(data)

# print([len(v) for v in data.values()])
```

#### 4.2 Printing the length of our columns in the dictionary

```python
print([len(v) for v in data.values()])
```

#### 4.3 Creating a DataFrame by use of the Pandas library

```python
# creating a dataframe and converting our data into a csv file
df = pd.DataFrame.from_dict(data, orient = 'index').transpose()
df
```

#### 4.4 Declaring our .CSV file name

```python
df.to_csv('linkedin_scraper.csv')
```
