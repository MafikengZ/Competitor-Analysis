```python
# Imports
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

# Query by Username
The code below will scrape for 100 tweets by a username then provide a CSV file with Pandas

```python
# Setting variables to be used below
maxTweets = 50000

# Creating list to append tweet data to
tweets_list1 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:udacity since:2021-01-01 until:2022-03-28').get_items()):
    if i>maxTweets:
        break
    tweets_list1.append([ tweet.user.username,tweet.date, tweet.id, tweet.content,tweet.likeCount,
                         tweet.media, tweet.source,tweet.replyCount,tweet.retweetCount])
```

```python
# Creating a dataframe from the tweets list above
tweets_df1 = pd.DataFrame(tweets_list1, columns=['Username','Datetime', 'Tweet Id', 'Text','likeCount',
                                                 'Media','Source','ReplyCount','RetweetCount'])

# Display first 5 entries from dataframe
tweets_df1.head(50000)
```

```python
#Export dataframe into a CSV
tweets_df1.to_csv('user_Twitter_data_udacity.csv', sep=',', index=False)
```
