```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/MafikengZ/Competitor-Analysis/main/Data/facebook/Facebook_GetSmarter.csv')
df.head()
```

```python
df_new = df[['message','Total_comments', 'total_reactions']]
df_new.head()
```

```python
df_new.plot(kind='scatter', x= 'Total_comments', y = 'total_reactions', color =  'red')
plt.show()
```
