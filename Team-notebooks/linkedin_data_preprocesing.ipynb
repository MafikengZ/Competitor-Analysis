The following codeblock is used to create a loop that iterates through our *_"Likes_Counts"_* columns in our moringa dataset and creating a list of the cleaned column and appending it into another empty list.

## Combining our *_CSV_* files into a single file known as combined_linked_data.csv

```python
os.chdir("csv")
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_linkedin_data.csv", index=False, encoding='utf-8-sig')
```
