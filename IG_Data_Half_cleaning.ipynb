```python
import numpy as np
import pandas as pd
import datetime
import pytz
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/MafikengZ/Competitor-Analysis/main/Data/instagram/instagram_posts.csv')
df.head()
```

```python
df['Id'] = df['Id'].astype('string')
df['post'] = df['post'].astype('string')
df['post_type'] = df['post_type'].astype('string')
df['tags'] = df['tags'].astype('string')
df['likes'] = df['likes'].astype('int64')
df['comments'] = df['comments'].astype('int64')
df['followers'] = df['followers'].astype('int64')
```

```python
df['views'] = df['views'].replace(np.nan, 0)
df['post'] = df['post'].replace(np.nan, 'No post')
df['post_type'] = df['post_type'].replace(np.nan, 'No post type')
df['tags'] = df['tags'].replace(np.nan, 'No tags')
```

```python
df['views'] = pd.to_numeric(df['views'], errors='coerce')
df
```

```python
missing_values = df.isna().sum()*100 / len(df)
missing_values
```

```python
df['views']= df['views'].astype('int64', errors='ignore')
```

```python
df.drop('sentiment', axis = 1, inplace=True)
df
```
